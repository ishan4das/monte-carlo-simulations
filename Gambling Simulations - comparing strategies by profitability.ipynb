```
import random
import matplotlib
import matplotlib.pyplot as plt
import time

from matplotlib import style
style.use('ggplot')  
plt.figure(figsize = (15 , 8))
def rollDice():
  roll = random.randint(1 , 100) 
  if ( roll == 100 ):
    return False

  elif (roll <= 50):
    return False

  elif ( 51 <= roll <= 99):
    return True

def doubler_bettor(funds, initial_wager, wager_count):
  value = funds #initial value, value is how much money he has
  wager = initial_wager #initial bet amount
  global broke_count_d
  global winners_d
  w_X = [] #for plotting
  v_Y = [] #for plotting

  currentWager = 1 #counter
  previousWager = 'win' #last bet win or lose
  previousWagerAmount = initial_wager #last bet amount


  while (currentWager <= wager_count):

    if (previousWager == 'win'):        #we had won previous role 
      wager = initial_wager
      if (rollDice()):
        value = value + wager
        w_X.append(currentWager)
        v_Y.append(value)
      else:
        value = value - wager
        previousWager = 'loss'
        previousWagerAmount = wager  # wager = initial wager
        w_X.append(currentWager)
        v_Y.append(value)
        if value < 0:
          broke_count_d += 1
          break



    elif( previousWager =='loss'):      #we had lost previous roll
      wager = previousWagerAmount * 2 #new bet amount

      if ((rollDice())):               #we won this roll
        if ((value - wager)<0):
          wager = value
        
        value = value + wager
        #Now         
        previousWager = 'win'            
        w_X.append(currentWager)
        v_Y.append(value)
      
      else:                           #we lost this roll 
        if ((value - wager)<0):
          wager = value
        
        value = value - wager
        #Now
        previousWager = 'loss'
        previousWagerAmount = wager
        w_X.append(currentWager)
        v_Y.append(value)

        if (value <= 0):
          broke_count_d += 1
          break

    currentWager += 1 

  plt.plot (w_X , v_Y , 'blue')
  if (value > funds):
    winners_d += 1 



def simple_bettor(funds , initial_wager, wager_count):
  value = funds #value is how much money he has
  wager = initial_wager #initial bet amount
  global broke_count_s
  global winners_s
  w_X = [] #for plotting
  v_Y = [] #for plotting

  currentWager = 1 #counter

  while currentWager <= wager_count:
    if (rollDice()) : 
      value = value + wager
      w_X.append(currentWager)
      v_Y.append(value)
    else:
      value = value - wager
      w_X.append(currentWager)
      v_Y.append(value)
      if (value <= 0 ):
        break

    currentWager = currentWager + 1
  
  if (value>funds):
    winners_s += 1

  if (value <= 0):
    value = 'broke'
    broke_count_s += 1
  plt.plot(w_X , v_Y , 'lime' )

       

def gambler_simulation(sampleSize_input , startingFunds_input , wagerSize_input , wagerCount_input):

  x = 0 
  while(x < sampleSize_input):
    simple_bettor(startingFunds_input , wagerSize_input , wagerCount_input)
    doubler_bettor(startingFunds_input , wagerSize_input , wagerCount_input)
    x = x + 1 

  print('broke gamblers % for simple bettor: ' , (float(broke_count_s)/float(sampleSize))*100)
  print('broke gamblers % for doubler bettor:' , (float(broke_count_d)/float(sampleSize))*100)

  print('profiting gamblers % for simple bettor', (float(winners_s)/float(sampleSize))*100)
  print('profiting gamblers % for doubler bettor' , (float(winners_d)/float(sampleSize))*100)
  plt.axhline(0 , color = 'r')
  plt.ylabel('Gambler\'s fund Balance ')
  plt.xlabel('Number of bets (Wager Count)')
  plt.show()

sampleSize = 100
startingFunds = 10000
wagerSize = 100 
wagerCount = 1000

broke_count_s = 0
broke_count_d = 0 
winners_s = 0 
winners_d = 0 

gambler_simulation(sampleSize , startingFunds, wagerSize , wagerCount)
```
