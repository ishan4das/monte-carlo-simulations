```
import random
import matplotlib
import matplotlib.pyplot as plt

import time

from matplotlib import style
style.use('ggplot')
plt.figure(figsize = (15 , 8))

def rollDice():
  roll = random.randint(1 , 100) 
  if ( roll == 100 ):
    return False

  elif (roll <= 50):
    return False

  elif ( 51 <= roll <= 99):
    return True

'''
the gambler who uses "doubler bettor" strategy doubles his bet size every time he loses. 

Basically, this is gambler believes in the "gambler's fallacy", so he thinks he is more likely to win the next wager if he loses in this wager. 
'''
def doubler_bettor(funds, initial_wager, wager_count):
  value = funds #initial value, value is how much money he has
  wager = initial_wager #initial bet amount

  w_X = [] #for plotting
  v_Y = [] #for plotting

  currentWager = 1 #counter
  previousWager = 'win' #last bet win or lose
  previousWagerAmount = initial_wager #last bet amount


  while (currentWager <= wager_count):

    if (previousWager == 'win'):        #we had won previous role 
      wager = initial_wager
      if (rollDice()):
        value = value + wager
        w_X.append(currentWager)
        v_Y.append(value)
      else:
        value = value - wager
        previousWager = 'loss'
        previousWagerAmount = wager  # wager = initial wager
        w_X.append(currentWager)
        v_Y.append(value)
        if value < 0:
          break



    elif( previousWager =='loss'):      #we had lost previous roll
      wager = previousWagerAmount * 2 #new bet amount

      if ((rollDice())):               #we won this roll
        value = value + wager
        #Now         
        previousWager = 'win'            
        w_X.append(currentWager)
        v_Y.append(value)
      
      else:                           #we lost this roll 
        value = value - wager
        #Now
        previousWager = 'loss'
        previousWagerAmount = wager
        w_X.append(currentWager)
        v_Y.append(value)

        if (value <= 0):
          break

    currentWager += 1 

  if (value < 0 ):
    print("Gambler finished with a debt of" , abs(value) , ' i.e funds = ', value)
  elif(value >= 0 ):
    print("Gambler finished with a profit of" , value - funds, ' i.e funds =', value)

  plt.plot (w_X , v_Y)

doubler_bettor(10000 , 10 , 1000)

plt.ylabel('Gambler\'s fund Balance ')
plt.xlabel('Number of bets (Wager Count)')
plt.show()
```

```
import random
import matplotlib
import matplotlib.pyplot as plt

import time

from matplotlib import style
style.use('ggplot')
plt.figure(figsize = (15 , 8))

def rollDice():
  roll = random.randint(1 , 100) 
  if ( roll == 100 ):
    return False

  elif (roll <= 50):
    return False

  elif ( 51 <= roll <= 99):
    return True

'''
the gambler who uses "doubler bettor" strategy doubles his bet size every time he loses. 

Basically, this is gambler believes in the "gambler's fallacy", so he thinks he is more likely to win the next wager if he loses in this wager. 
'''
def doubler_bettor(funds, initial_wager, wager_count):
  value = funds #initial value, value is how much money he has
  wager = initial_wager #initial bet amount
  global broke_count 

  w_X = [] #for plotting
  v_Y = [] #for plotting

  currentWager = 1 #counter
  previousWager = 'win' #last bet win or lose
  previousWagerAmount = initial_wager #last bet amount


  while (currentWager <= wager_count):

    if (previousWager == 'win'):        #we had won previous role 
      wager = initial_wager
      if (rollDice()):
        value = value + wager
        w_X.append(currentWager)
        v_Y.append(value)
      else:
        value = value - wager
        previousWager = 'loss'
        previousWagerAmount = wager  # wager = initial wager
        w_X.append(currentWager)
        v_Y.append(value)
        if value < 0:
          broke_count = broke_count + 1
          break



    elif( previousWager =='loss'):      #we had lost previous roll
      wager = previousWagerAmount * 2 #new bet amount

      if ((rollDice())):               #we won this roll
        value = value + wager
        #Now         
        previousWager = 'win'            
        w_X.append(currentWager)
        v_Y.append(value)
      
      else:                           #we lost this roll 
        value = value - wager
        #Now
        previousWager = 'loss'
        previousWagerAmount = wager
        w_X.append(currentWager)
        v_Y.append(value)

        if (value <= 0):
          broke_count = broke_count + 1
          break

    currentWager += 1 
  plt.plot (w_X , v_Y)


xx = 0 #counter
broke_count = 0 
while (xx < 1000):
  doubler_bettor(10000 , 10 , 100)
  xx = xx + 1


plt.ylabel('Gambler\'s fund Balance ')
plt.xlabel('Number of bets (Wager Count)')

broke_rate = (float(broke_count)/ float(xx))*100
print( 'broke % = ' , broke_rate )
print('survival %' , 100 - broke_rate)
plt.axhline( 0 , color = 'r')
plt.show()
```
