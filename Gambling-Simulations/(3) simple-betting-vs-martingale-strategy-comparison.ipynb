```
import random
import matplotlib
import matplotlib.pyplot as plt

import time

from matplotlib import style
style.use('ggplot')
plt.figure(figsize = (15 , 8))

def rollDice():
  roll = random.randint(1 , 100) 
  if ( roll == 100 ):
    return False

  elif (roll <= 50):
    return False

  elif ( 51 <= roll <= 99):
    return True

'''
the gambler who uses "doubler bettor" strategy doubles his bet size every time he loses. 

Basically, this is gambler believes in the "gambler's fallacy", so he thinks he is more likely to win the next wager if he loses in this wager. 
'''
def doubler_bettor(funds, initial_wager, wager_count):
  value = funds #initial value, value is how much money he has
  wager = initial_wager #initial bet amount

  w_X = [] #for plotting
  v_Y = [] #for plotting

  currentWager = 1 #counter
  previousWager = 'win' #last bet win or lose
  previousWagerAmount = initial_wager #last bet amount


  while (currentWager <= wager_count):

    if (previousWager == 'win'):        #we had won previous role 
      wager = initial_wager
      if (rollDice()):
        value = value + wager
        w_X.append(currentWager)
        v_Y.append(value)
      else:
        value = value - wager
        previousWager = 'loss'
        previousWagerAmount = wager  # wager = initial wager
        w_X.append(currentWager)
        v_Y.append(value)
        if value < 0:
          break



    elif( previousWager =='loss'):      #we had lost previous roll
      wager = previousWagerAmount * 2 #new bet amount

      if ((rollDice())):               #we won this roll
        if ((value - wager)<0):
          wager = value
        value = value + wager
        #Now         
        previousWager = 'win'            
        w_X.append(currentWager)
        v_Y.append(value)
      
      else:                           #we lost this roll 
        if ((value - wager)<0):
          wager = value
        value = value - wager
        #Now
        previousWager = 'loss'
        previousWagerAmount = wager
        w_X.append(currentWager)
        v_Y.append(value)

        if (value <= 0):
          break

    currentWager += 1 
  plt.plot (w_X , v_Y)



def simple_bettor(funds , initial_wager, wager_count):
  value = funds #value is how much money he has
  wager = initial_wager #initial bet amount
  global broke_count
  w_X = [] #for plotting
  v_Y = [] #for plotting

  currentWager = 1 #counter

  while currentWager <= wager_count:
    if (rollDice()) : 
      value = value + wager
      w_X.append(currentWager)
      v_Y.append(value)
    else:
      value = value - wager
      w_X.append(currentWager)
      v_Y.append(value)
      if (value <= 0 ):
        break

    currentWager = currentWager + 1
  
  if (value <= 0):
    value = 'broke'
    broke_count += 1
  plt.plot(w_X , v_Y)    


x = 0 
broke_count = 0 
while(x<1000):
  simple_bettor(10000 , 100 , 100)
  x = x + 1 


broke_rate = (float(broke_count)/ float(x))*100
print( 'broke % = ' , broke_rate )
print('survival %' , 100 - broke_rate)


plt.ylabel('Gambler\'s fund Balance ')
plt.xlabel('Number of bets (Wager Count)')
plt.show()
```

```
import random
import matplotlib
import matplotlib.pyplot as plt
import time

from matplotlib import style
style.use('ggplot')
plt.figure(figsize = (15 , 8))

def rollDice():
  roll = random.randint(1 , 100) 
  if ( roll == 100 ):
    return False

  elif (roll <= 50):
    return False

  elif ( 51 <= roll <= 99):
    return True

def doubler_bettor(funds, initial_wager, wager_count):
  value = funds #initial value, value is how much money he has
  wager = initial_wager #initial bet amount
  global broke_count_d
  w_X = [] #for plotting
  v_Y = [] #for plotting

  currentWager = 1 #counter
  previousWager = 'win' #last bet win or lose
  previousWagerAmount = initial_wager #last bet amount


  while (currentWager <= wager_count):

    if (previousWager == 'win'):        #we had won previous role 
      wager = initial_wager
      if (rollDice()):
        value = value + wager
        w_X.append(currentWager)
        v_Y.append(value)
      else:
        value = value - wager
        previousWager = 'loss'
        previousWagerAmount = wager  # wager = initial wager
        w_X.append(currentWager)
        v_Y.append(value)
        if value < 0:
          broke_count_d += 1
          break



    elif( previousWager =='loss'):      #we had lost previous roll
      wager = previousWagerAmount * 2 #new bet amount

      if ((rollDice())):               #we won this roll
        if ((value - wager)<0):
          wager = value
        
        value = value + wager
        #Now         
        previousWager = 'win'            
        w_X.append(currentWager)
        v_Y.append(value)
      
      else:                           #we lost this roll 
        if ((value - wager)<0):
          wager = value
        
        value = value - wager
        #Now
        previousWager = 'loss'
        previousWagerAmount = wager
        w_X.append(currentWager)
        v_Y.append(value)

        if (value <= 0):
          broke_count_d += 1
          break

    currentWager += 1 
  plt.plot (w_X , v_Y , 'blue')



def simple_bettor(funds , initial_wager, wager_count):
  value = funds #value is how much money he has
  wager = initial_wager #initial bet amount
  global broke_count_s
  w_X = [] #for plotting
  v_Y = [] #for plotting

  currentWager = 1 #counter

  while currentWager <= wager_count:
    if (rollDice()) : 
      value = value + wager
      w_X.append(currentWager)
      v_Y.append(value)
    else:
      value = value - wager
      w_X.append(currentWager)
      v_Y.append(value)
      if (value <= 0 ):
        break

    currentWager = currentWager + 1
  
  if (value <= 0):
    value = 'broke'
    broke_count_s += 1
  plt.plot(w_X , v_Y , 'lime' )    

sampleSize = 1000
startingFunds = 10000
wagerSize = 100 
wagerCount = 1000

x = 0 
broke_count_s = 0
broke_count_d = 0 

while(x < sampleSize):
  simple_bettor(startingFunds , wagerSize , wagerCount)
  doubler_bettor(startingFunds , wagerSize , wagerCount)
  x = x + 1 

broke_rate_s = (float(broke_count_s)/ float(x))*100
print( 'broke % for simple bettor = ' , broke_rate_s )
print('survival % for simple bettor' , 100 - broke_rate_s)

broke_rate_d = (float(broke_count_d)/ float(x))*100
print( 'broke % for doubler bettor = ' , broke_rate_d )
print('survival % for doubler bettor' , 100 - broke_rate_d)

plt.ylabel('Gambler\'s fund Balance ')
plt.xlabel('Number of bets (Wager Count)')
plt.show()
```

```
def gambler_simulation(sampleSize_input , startingFunds_input , wagerSize_input , wagerCount_input):

  x = 0 
  while(x < sampleSize_input):
    simple_bettor(startingFunds_input , wagerSize_input , wagerCount_input)
    doubler_bettor(startingFunds_input , wagerSize_input , wagerCount_input)
    x = x + 1 

  broke_rate_s = (float(broke_count_s)/ float(x))*100
  print( 'broke % for simple bettor = ' , broke_rate_s )
  print('survival % for simple bettor' , 100 - broke_rate_s)

  broke_rate_d = (float(broke_count_d)/ float(x))*100
  print( 'broke % for doubler bettor = ' , broke_rate_d )
  print('survival % for doubler bettor' , 100 - broke_rate_d)

  plt.ylabel('Gambler\'s fund Balance ')
  plt.xlabel('Number of bets (Wager Count)')
  plt.show()

broke_count_s = 0 #cannot change/assign global variable = 0 from inside the function, so have to assign outside every time
broke_count_d = 0

sampleSize = 1000
startingFunds = 10000
wagerSize = 100 
wagerCount = 1000  
gambler_simulation(1000 , 10000 , 100 , 1000)
```
